```python
w1 = 50      # Frequência da primeira componente do sinal
w2 = 100      # Frequência da segunda componente do sinal    
w3 = 500      # Frequência da terceira componente do sinal 
t=np.arange(0, 0.1, 10e-6) # Vetor de tempo
x = sin(2*pi*w1*t)+sin(2*pi*w2*t)+sin(2*pi*w3*t) # Sinal x contínuo no tempo

plt.figure('Figura 1')
plt.title('Sinal $x(t)$ contínuo')
plt.xlabel('Tempo') 
plt.ylabel('x(t)')
plt.plot(t,x)
plt.xlim(0, 0.1)
```

2

```python
fs = 1000
N=1001                           
n= np.arange(0, N-1,1)          # Vetor de pontos
Ts=1/fs                         # Periodo de amostragem
k= np.arange( 0, len(n)*Ts, Ts) # Vetor de tempo discreto

xd = sin((2*pi*w1/fs)*n) + sin((2*pi*w2/fs)*n) + sin((2*pi*w3/fs)*n) # Sinal x discreto no tempo (amostrado)
f = interpolate.interp1d(k,xd,kind='zero')                           # Função interpolada em ordem zero
xn = f(t)                                                            # Sinal segurado em ordem zero

plt.figure('Figura 1')
plt.title('Sinal $x(t)$ com taxa de amostragem de $1kHz$')
plt.plot(k, xd,'o') 
plt.xlabel('k')
plt.ylabel('x[k]')
plt.plot(t,xn)
plt.xlim(0, 0.1)
plt.legend(['Sinal discreto', 'Sinal $x(t)$'],fontsize='small',loc='lower right')
```

3

```python
plt.figure('Figura 1')
plt.subplot(2,1,1)
plt.title('Sinal $x(t)$ contínuo e discreto e amostragem de $1kHz$')
plt.plot(t,x,'r') 
plt.plot(t,xn,'b')
plt.xlabel('Tempo') 
plt.ylabel('x(t)')
plt.xlim(0, 0.1)
plt.legend(['Sinal Contínuo', 'Sinal amostrado'],fontsize='small',loc='lower right')

y= sp.fft.fft(xd)           # Transformada rápida de Fourier  
fs= sp.fft.fftfreq(N-1, Ts) # Eixo de frequências entre -fs/2 a fs/2 
plt.figure('Figura 2')
plt.subplot(2,1,2)
plt.title('Espectro de frequências de $x(t)$')
plt.plot(fs,abs(y), 'b')    # Espectro de frequências
plt.xlabel('Freq. ($Hz$)')
plt.ylabel('Amplitude')
```

```python
ti1, yi1  = ct.impulse_response(G)
ti2, yi2  = ct.impulse_response(Gd1)
ti3, yi3  = ct.impulse_response(Gd2)
ti4, yi4  = ct.impulse_response(Gd3)

plt.figure(1)
plt.title("Resposta ao impulso")
plt.plot(ti1.T, yi1.T,'r',label='Sistema contínuo')
plt.plot(ti2.T, yi2.T,'g',label='Sistema com Ts = 0,045s')
plt.plot(ti3.T, yi3.T,'b',label='Sistema com Ts = 0,0045s')
plt.plot(ti4.T, yi4.T,'y',label='Sistema com Ts = 0,45s')
plt.xlim(0,8)
plt.legend()
plt.grid()


t1, y1  = ct.step_response(G)
t2, y2  = ct.step_response(Gd1)
t3, y3  = ct.step_response(Gd2)
t4, y4  = ct.step_response(Gd3)
t5, y5  = ct.step_response(Gdt1)
t6, y6  = ct.step_response(Gdt2)
t7, y7  = ct.step_response(Gdt3)

plt.figure(2)
plt.subplot(2,1,1)
plt.title("Resposta ao degrau")
plt.plot(t1.T, y1.T,'r',label='Sistema contínuo')
plt.plot(t2.T, y2.T,'g',label='Sistema com Ts = 0,045s')
plt.plot(t3.T, y3.T,'b',label='Sistema com Ts = 0,0045s')
plt.plot(t4.T, y4.T,'y',label='Sistema com Ts = 0,45s')
plt.xlim(0,8)
plt.legend()
plt.grid()

plt.subplot(2,1,2)
plt.plot(t1.T, y1.T,'r',label='Sistema contínuo')
plt.plot(t5.T, y5.T,'C9',label='Sistema com Ts = 0,045s')
plt.plot(t6.T, y6.T,'C5',label='Sistema com Ts = 0,0045s')
plt.plot(t7.T, y7.T,'C6',label='Sistema com Ts = 0,45s')
plt.xlim(0,8)
plt.legend(fontsize='small')
plt.grid()

```
